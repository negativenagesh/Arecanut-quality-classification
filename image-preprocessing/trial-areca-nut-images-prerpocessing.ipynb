```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exampleimg=cv2.imread("example-image.jpg")
cvimage=cv2.cvtColor(exampleimg,cv2.COLOR_BGR2GRAY)

cv2.namedWindow("window", cv2.WINDOW_NORMAL)
cv2.resizeWindow("window", 400,512 )

print(cvimage.shape)

cv2.imshow("window",cvimage)
cv2.waitKey(0)
```

# Blurring (Gaussian blur)

```python
blur = cv2.GaussianBlur(exampleimg,(5,5),0)
blur
```

```python
blur = cv2.GaussianBlur(exampleimg,(5,5),0)
 
plt.subplot(121),plt.imshow(exampleimg),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
